```python
from curve_fitting.data_utils import plot_results
```

```python
data_path = 'CFs_for_segments/SPE_carb/'
```

```python
files = os.listdir(data_path)
```

```python
import numpy as np
```

```python
results = dict()
for fname in files:
    results_stack = read_data(data_path, fname)
    for k, v in results_stack.items():
        if k not in results:
            results[k] = dict()
        v = np.array(v)
        results[k][fname] = v / v[0] 
```

```python
results.keys()
```

```python
plot_results(results['S2Y'], labels=results['S2Y'].keys(), x_down=0, x_up=1)
```

```python
plot_results(results['S2X'], labels=results['S2X'].keys(), x_down=0, x_up=1)
```
